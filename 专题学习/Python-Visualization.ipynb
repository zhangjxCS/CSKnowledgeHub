# Python Visualization

This is a tutorial to make visualization using Python. There are two kinds of packages for visualization in Python, which are Matplotlib and Seaborn
1. Matplotlib is used to create statical interferences and plotting 2D graphs of arrays
2. Seaborn is built on the roof of Matplotlib and can help in visualizing univariate and bivariate data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

# Continuous Univariate Visualization

For continuous univariate visualization, there are usually three types of graphs, which are histogram, boxplot, and QQ plot 

```python
iris = sns.load_dataset('iris')
iris.head()
```

## Histogram

Histogram is used to visualize the continuous variables. There are three types of histogram, frequency, cumulative frequency and density. Histogram is good for single variable distribution visualization.

Sometimes, we draw a density graph on a histogram. A density plot is a representation of the distribution of a numeric variable. It uses a kernel density estimate to show the probability density function of the variable. It is a smoothed version of the histogram and is used in the same concept.

```python
plt.hist(x=iris['sepal_length'],      # array or sequence 
         bins=10,                     # number of bins or sequence of bin edges
         range=None,                  # lower and upper range of the bins
         density=False,               # draw and return a probability density
         weights=None,                # an array of weights for data
         cumulative=False,            # cumulative histogram
         bottom=None,                 # shift of bottom
         histtype='bar',              # bar, barstacked, step, stepfilled
         align='mid',                 # horizontal alignment of the histogram bars
         orientation='vertical',      # vertical of horizontal bar chart
         rwidth=None,                 # relative width of the bars as a fraction of the bin width
         log=False,                   # If True, the histogram axis will be set to a log scale
         color=None,                  # color of bins
         label=None,                  # sequence of strings to match multiple datasets
         stacked=False)               # stacked histogram
plt.title('Sepal Length for Iris Data')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('count')
plt.show()
```

```python
sns.histplot(data=iris['sepal_length'],      # data
             x=None, y=None,                 # positions on the x and y axes
             hue=None,                       # color of plot elements
             weights=None,                   # weights of data points
             stat='count',                   # y axis, count, frequency, probability, percent, density
             bins='auto',                    # number of bins
             binwidth=None,                  # width of each bin
             binrange=None,                  # bin edges
             discrete=None,                  # binwidth=1
             cumulative=False,               # cumulative histogram
             kde=True,                       # kernel density plot
             color=None,                     # color of plot
             log_scale=None,                 # log scale of axis
             legend=False)                   # legend
plt.title('Sepal Length for Iris Data')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('count')
```

## Boxplot

Boxplot is used to visualize the five number summary of data, which are minimum, first quartile, median, third quartile, and maximum. Compared to Histogram, boxplot is easy to compare difference between groups.

```python
plt.boxplot(x=iris.iloc[:, :4],      # data
            notch=None,              # notched box plot or rectangular box plot
            vert=None,               # vertical or horizontal boxes
            whis=None,               # position of whiskers
            positions=None,          # positions of boxes
            widths=None,             # width of the boxes
            labels=iris.columns[:4], # labels of boxes
            meanline=True,           # render the mean as a line
            showmeans=True)          # show means
plt.title('Sepal Statistics for Iris Data')
plt.xlabel('Variables')
plt.ylabel('Length (cm)')
plt.show()
```

## Q-Q plot

Qâ€“Q (quantile-quantile) plot is a probability plot, which is a graphical method for comparing two probability distributions by plotting their quantiles against each other. Q-Q plot is used to compare whether one variable is normally distributed or not.

```python
from scipy import stats

plot = stats.probplot(iris['sepal_length'], plot=sns.mpl.pyplot)
```

# Categorical Univariate Visualization

We usually use bar chart to visualize univariate categorical variable

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

## Bar Chart

A bar chart is a chart that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. It is usually used to visualize   categorical variables

```python
sns.countplot(x='class', y=None,  # inputs
              hue=None,           # inputs for groups
              data=titanic,       # dataset
              order=None,         # order to plot the categorical levels
              hue_order=None,     # order to plot the hue levels
              orient=None,        # vertical or horizontal graphs
              color=None,         # color for all elements
              palette=None,       # colors to use for different levels
              saturation=0.75,    # proportion of the original saturation to draw colors at
              dodge=True,         # whether elements should be shifted when hue is used
              ax=None)            # axes to draw
```

# Continuous Multivariate Visualization

For continuous multivariate visualization, we usually use scatter plot and 2D density plot.

## Scatter Plot

A scatter plot is used to display values for two variables for a set of data. The scatter plot is easy to view the dependency of two variables.

```python
plt.scatter(x=iris['sepal_width'],   # data
            y=iris['sepal_length'],  # data
            s=10,                    # marker size
            c=None,                  # marker colors
            marker=None,             # market style
            alpha=None,              # alpha blend
            linewidths=None,         # linewidths of the marker edges
            edgecolors=None,         # edge color of the marker
            plotnonfinite=False)     # whether to plot points with non finite c
plt.title('Sepal Width and Length of Iris Data')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.show()
```

## 2D Density Plot

A 2D density plot is usedful to study the relationship between two numeric variables.

```python
sns.kdeplot(x=iris['sepal_width'],  # x variable
            y=iris['sepal_length'], # y variable
            cumulative=False,       # cumulative density
            hue=None,               # plot group
            color=None,             # plot color
            fill=None,              # fill or not
            warn_singular=True)     # warning of 0 variance
```

# Categorical Multivariate Visualization

For categorical multivariate visualization, we usually use bar chart of mosaic plot.

## Bar Chart

Group bar chart is a good way to visualize the multivariate categorical variables

```python
sns.countplot(x='class', y=None,  # inputs
              hue='sex',          # inputs for groups
              data=titanic,       # dataset
              order=None,         # order to plot the categorical levels
              hue_order=None,     # order to plot the hue levels
              orient=None,        # vertical or horizontal graphs
              color=None,         # color for all elements
              palette=None,       # colors to use for different levels
              saturation=0.75,    # proportion of the original saturation to draw colors at
              dodge=True,         # whether elements should be shifted when hue is used
              ax=None)            # axes to draw
```

## Mosaic Plot

A mosaic plot is a graphical method for visualizing data from two or more qualitative variables. Mosaic plot is easy to view the relationship between two categorical variables.

```python
from statsmodels.graphics.mosaicplot import mosaic

mosaic(data=titanic,                   # data
       index=['class', 'survived'],    # index
       ax=None,                        # the graph where display the mosaic
       horizontal=True,                # starting direction of the split
       gap=0.01,                       # list of gap to be applied on each subdivision
       labelizer=None,                 # text to display at the center of each tile
       title='Survival rate for different classes', 
       statistic=None,                 # statistical model to give colors to the plot
       axes_label=True,                # show the name of each value of each category
       label_rotation=0.0)             # rotation of axis label
```
